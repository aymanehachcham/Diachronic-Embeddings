```python
import pandas as pd
```

```python
df_80 = pd.read_csv('../articles_raw_data/news_1980.csv')
df_85 = pd.read_csv('../articles_raw_data/news_1985.csv')
df_90 = pd.read_csv('../articles_raw_data/news_1990.csv')
df_95 = pd.read_csv('../articles_raw_data/news_1995.csv')
df_00 = pd.read_csv('../articles_raw_data/news_2000.csv')
df_05 = pd.read_csv('../articles_raw_data/news_2005.csv')
df_10 = pd.read_csv('../articles_raw_data/news_2010.csv')
```

```python
df_15 = pd.read_csv('../articles_raw_data/news_2015.csv')
```

```python
all_words = []

all_docs = pd.concat([df_80['text'], df_85['text'], df_90['text'], df_95['text'], df_00['text'], df_05['text'], df_10['text'], df_15['text']]) 

for doc in all_docs:
    words = str(doc).split()

    all_words += words
```

# Getting the list of target words:

```python
from collections import Counter
```

```python
all_words = list(filter(lambda x: len(x) > 4, all_words))
```

```python
def most_common_tokens(n, words):

    counter = Counter(words)

    return counter.most_common(n)
```

```python
most_common_words = most_common_tokens(200, all_words)
```

```python
most_comon = []
for i in range(200):
    most_comon.append(most_common_words[i][0])
```
