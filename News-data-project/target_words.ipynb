```python
import pandas as pd
from collections import Counter
```

```python
df_80 = pd.read_csv('data_csv/news_1980.csv')
```

```python
df_80
```

```python
df_80 = df_80.dropna()
```

```python
df_80.info()
```

```python

words_80 =  []

for text in df_80['text']:
    words_80 += str(text).split()
```

```python
word_freq_80 = []
for k,v in Counter(words_80).items():
    if v > 100:
        word_freq_80.append(k)
```

```python
word_freq_80 = list(filter(lambda x: len(x) > 1, word_freq_80))
```

```python
word_freq_80
```
